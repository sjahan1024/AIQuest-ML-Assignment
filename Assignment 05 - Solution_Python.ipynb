# 1. Normalization

```python
import pandas as pd
df = pd.read_excel("Assignment 05 - Data set.xlsx")
df
```

```python
# Normalization calculate manually
df['price_norm_manual'] = (df.Price - df.Price.min()) / (df.Price.max() - df.Price.min())
df
```

```python
# Normalization calculate sklearn
from sklearn.preprocessing import MinMaxScaler
mmx = MinMaxScaler()
df['price_norm_sklearn'] = mmx.fit_transform(df[['Price']])
df
```
